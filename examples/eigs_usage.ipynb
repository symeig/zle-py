```python
# symbolic eigenvalue verification 
# makes each variable a random value and does a general eigenvalue calculation, substitutes the values into the matrix, and checks that
# the values match up
def verify_numerically(symsol,A):
    for i in range(2):
        rvals=np.random.rand(n)*2-1
        At = np.array(A.subs({k: v for k, v in zip(s, rvals)})).astype(np.float64)#.tolist())

        w,v=np.linalg.eig(At)
        idx=np.argsort(w)
        w=np.sort(w)
        v=v[idx]
        
        sw = np.array(sp.Matrix(symsol).subs({k: v for k, v in zip(s, rvals)})).astype(np.float64)
        sw = np.sort([np.sum(sw[j,:]) for j in range(n)])
        if (np.array_equal(np.round(w,3),np.round(sw,3))==False):
            print(w,'\n\n',sw)
            return False
    return True
```
